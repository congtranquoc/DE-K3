```python
ecomerce_csv = pd.read_csv('Ecommerce_Purchases.csv', usecols=['Job', 'Purchase Price'])
```

```python
ecomerce_csv_sum = ecomerce_csv.groupby('Job',group_keys=False,dropna=False).agg({'Purchase Price': 'sum'}).sort_values(by='Purchase Price', ascending=False).reset_index()
```

```python
ecomerce_csv_sum.head(5)
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
dt = ecomerce_csv_sum.head(5)
jobs = list(dt["Job"])
total_purchar = list(dt["Purchase Price"])
# Tạo một dãy màu
colors = list(mcolors.TABLEAU_COLORS.values()) 
fig = plt.figure(figsize = (10, 5))
# sns.barplot(x="Job", y="Purchase Price", data=dt)
plt.bar(jobs, total_purchar, color=colors, width = 0.8)
plt.xlabel('Profession')
plt.ylabel('Total Purchase')
plt.xticks(rotation=45)
plt.show()
```

```python
df_mean = ecomerce_csv.groupby('Job')['Purchase Price'].mean().reset_index()
```

```python
df_mean = df_mean.sort_values(by='Purchase Price', ascending=False)
```

```python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
dt = df_mean.head(5)
jobs = list(dt["Job"])
total_purchar = list(dt["Purchase Price"])
# Tạo một dãy màu
colors = list(mcolors.TABLEAU_COLORS.values()) 
fig = plt.figure(figsize = (10, 5))
# sns.barplot(x="Job", y="Purchase Price", data=dt)
plt.bar(jobs, total_purchar, color=colors, width = 0.8)
plt.xlabel('Profession')
plt.ylabel('Total Purchase')
plt.xticks(rotation=45)
plt.show()
```

```python
purchase_value  = pd.read_csv('Ecommerce_Purchases.csv', usecols=['Job', 'Purchase Price', 'Browser Info'])
purchase_value['Browser Info'] = purchase_value['Browser Info'].str.split('/').str[0].str.strip()
# Group by Job and Browser Info and sum Purchase Price
grouped = purchase_value.groupby(['Job', 'Browser Info'])['Purchase Price'].sum().unstack().reindex(columns=['Opera', 'Mozilla']).reset_index()

list_merger = pd.merge(ecomerce_csv_sum, grouped, on='Job', how='left')

list_merger.head(5)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = list_merger.head(5)
df.sort_values(by='Purchase Price', ascending=False)
jobs = df['Job']
operas = df['Opera']
mozillas = df['Mozilla']
# Tạo một mảng numpy chứa các chỉ số cho các vị trí của các job trên trục y
y_indexes = np.arange(len(jobs))

# Vẽ biểu đồ
fig, ax = plt.subplots()
ax.barh(y_indexes + 0.4, mozillas, height=0.4, color='#1f77b4', label='Mozilla')
ax.barh(y_indexes, operas, height=0.4, color='#ff7f0e', label='Opera')


# Đặt tên cho trục y và x
ax.set_yticks(y_indexes + 0.4 / 2)
ax.set_yticklabels(jobs)
ax.set_xticklabels([])
ax.set_xlabel('Total Count Of Purchases')
ax.set_title('Browser Prefereces of largest job groups')

# Thêm chú thích
ax.legend(['Mozilla', 'Opera'], loc='center right',bbox_to_anchor=(1.25, 0.5))

plt.show()

```

```python
location_csv = pd.read_csv('Ecommerce_Purchases.csv', usecols=['Lot', 'Purchase Price', 'AM or PM'])
location_csv['Lot'] = location_csv['Lot'].str.split().str[1].str.strip()
location_csv['Lot'] = location_csv['Lot'].str.upper()
group_location = location_csv.groupby(['Lot', 'AM or PM']).agg({'Purchase Price': 'sum'}).unstack().reset_index()
group_location['Total'] =  group_location.select_dtypes(include=np.number).sum(axis=1)
group_location.sort_values(by=([('Total', '')]), ascending=False)
group_location
```

```python
df = group_location.sort_values(by=([('Total', '')]), ascending=False)
df = df.head(5)
lotcations = df['Lot']
am = df['Purchase Price', 'AM']
pm = df['Purchase Price', 'PM']

# Tạo một mảng numpy chứa các chỉ số cho các vị trí của các job trên trục y
y_indexes = np.arange(len(lotcations))

# Vẽ biểu đồ
fig, ax = plt.subplots()
ax.barh(y_indexes + 0.4, am, height=0.4, color='#1f77b4', label='AM')
ax.barh(y_indexes, pm, height=0.4, color='#ff7f0e', label='PM')


# Đặt tên cho trục y và x
ax.set_yticks(y_indexes + 0.4 / 2)
ax.set_yticklabels(lotcations)
ax.set_xticklabels([])
ax.set_xlabel('Total Count Of Purchases')
ax.set_ylabel('Top ordering states')
ax.set_title('Most purchases done by State (locations) and time')

# Thêm chú thích
ax.legend(['AM', 'PM'], loc='lower right')

plt.show()
```

```python
cc_provider_csv = pd.read_csv('Ecommerce_Purchases.csv', usecols=['CC Provider', 'Purchase Price', 'AM or PM'])

cc_mapping = {
    'JCB 16 digit': 'JCB',
    'JCB 15 digit': 'JCB',
    'VISA 13 digit': 'VISA',
    'VISA 16 digit': 'VISA'
}
cc_provider_csv['CC Provider'] = cc_provider_csv['CC Provider'].replace(cc_mapping)
cc_provider_csv = cc_provider_csv.groupby(['CC Provider', 'AM or PM']).agg({'Purchase Price': ['sum'], 'AM or PM': ['count']}).unstack().reset_index()
cc_provider_csv
```

```python
df = cc_provider_csv
providers = df['CC Provider']
purchase_am = df['Purchase Price','sum','AM']
purchase_pm = df['Purchase Price','sum','PM']
count_am = df['AM or PM','count','AM']
count_pm = df['AM or PM','count','PM']

y_indexes = np.arange(len(providers))

# Vẽ biểu đồ
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))


ax1.barh(y_indexes + 0.4, count_am, height=0.2, color='#1f77b4', label='AM')
ax1.barh(y_indexes, count_pm, height=0.2, color='#ff7f0e', label='PM')
ax1.set_yticks(y_indexes + 0.4 / 2)
ax1.set_yticklabels(providers)
ax1.set_xticklabels([])
ax1.set_xlabel('Numbers of Orders')
ax1.set_ylabel('CC Provider')
ax1.set_title('CC Provider usage count at different times')
ax1.legend(['AM', 'PM'], loc='upper right')



# Biểu đồ cho tổng giá trị đơn hàng
ax2.barh(y_indexes - 0.2, purchase_am, height=0.2, color='#9467bd', label='AM')
ax2.barh(y_indexes - 0.4, purchase_pm, height=0.2, color='#8c564b', label='PM')
ax2.set_yticks(y_indexes - 0.3)
ax2.set_yticklabels(providers)
ax2.set_xticklabels([])
ax2.set_xlabel('Purchase value')
ax2.set_title('CC Provider usage value at different times')

ax2.legend(['AM Total', 'PM Total'], loc='upper right')
plt.show()
```
