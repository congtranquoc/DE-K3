```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colors as mcolors
# Đọc dữ liệu từ file CSV
data = pd.read_csv('data.csv')

brand_counts = data['Branding'].value_counts()

brand_counts = brand_counts.sort_index()

colors = list(mcolors.TABLEAU_COLORS.values())

plt.figure(figsize=(30, 15))
bars = plt.bar(brand_counts.index, brand_counts.values, color=colors[:len(brand_counts)])

plt.title('Số lượng sản phẩm của mỗi hãng')
plt.xlabel('Hãng')
plt.ylabel('Số lượng')
plt.xticks(rotation=45)
plt.grid(True)

# Thêm chú thích cho mỗi thanh bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, str(int(height)), ha='center', va='bottom')

plt.show()
```

```python
import seaborn as sns
import numpy as np


plt.figure(figsize=(10, 6))
plt.hist(data['Price'], bins=20, color='#4c72b0', edgecolor='black')
plt.title('Phân bố giá của các sản phẩm', fontsize=14)
plt.xlabel('Giá', fontsize=12)
plt.ylabel('Số lượng', fontsize=12)
plt.tight_layout()
plt.show()
```

```python

data['Price'] = pd.to_numeric(data['Price'], errors='coerce')

plt.figure(figsize=(12, 8))  # Thiết lập kích thước hình vẽ
data.boxplot(column='Price', by='Branding', patch_artist=True, showfliers=False)
plt.title('Phân bố giá sản phẩm theo hãng')
plt.xlabel('Hãng')
plt.ylabel('Giá')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python

data['Price'] = pd.to_numeric(data['Price'], errors='coerce')
data['Rating'] = pd.to_numeric(data['Rating'], errors='coerce')

plt.figure(figsize=(10, 6))
plt.scatter(data['Price'], data['Rating'])
plt.title('Mối liên hệ giữa giá sản phẩm và rating')
plt.xlabel('Giá sản phẩm')
plt.ylabel('Rating')
plt.show()
```
