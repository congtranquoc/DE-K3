```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import colors as mcolors
# c d liu t file CSV
data = pd.read_csv('data.csv')

brand_counts = data['Branding'].value_counts()

brand_counts = brand_counts.sort_index()

colors = list(mcolors.TABLEAU_COLORS.values())

plt.figure(figsize=(30, 15))
bars = plt.bar(brand_counts.index, brand_counts.values, color=colors[:len(brand_counts)])

plt.title('S lng sn phm ca mi hng')
plt.xlabel('Hng')
plt.ylabel('S lng')
plt.xticks(rotation=45)
plt.grid(True)

# Thm ch thch cho mi thanh bar
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width() / 2, height, str(int(height)), ha='center', va='bottom')

plt.show()
```

```python

data['Price'] = pd.to_numeric(data['Price'], errors='coerce')

plt.figure(figsize=(12, 8))  # Thit lp kch thc hnh v
data.boxplot(column='Price', by='Branding', patch_artist=True, showfliers=False)
plt.title('Phn b gi sn phm theo hng')
plt.xlabel('Hng')
plt.ylabel('Gi')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
